```python
# Q. create a new list with even numbers from existing list 

# way 2

num_list=[24,12,13,14,15,36,17,88,79,101]

# List Comprehension example
# list1= [return value         for loop        condition ]



         #3               #1                     #2
list1= [x         for x in num_list         if x%2==0 ]    # adding space for understanding 

# no ":" needed in this , because no indentation needed , all in single line

# Here , inside list comprehension, the flow takes place in the following order: 
# 1. the numlist is iterated using the for loop. We fetch the values from the numlist via loop iteration
# 2. then it checks the condition whatever we want, 
# 3. and then it returns the value depending on then condition and puts it in the x in the third step. 



list1= [x for x in num_list if x%2==0 ] # just one space needed
list1
```

```python
#file handling : 
# while handling different types of files, they are done by different ways. 
#  The current method is only for text files. 

#text:

f=open("filename","mode") # filename must be with location 



text:  #type of data has different modes 

'r' or 'rt'  : read : default mode. # In R/T, R=read, T=text; can also be written just R, which means the same.  
'w' or 'wt' : write : if file exists : override , if file not exists : create
'a' or 'at' : append : if file exists : append , if file not exists : create
'r+' or 'rt+' : read and write
'w+' or 'wt+' : write and read
'a+' or 'at+' : append and read
    
    


byte:  #type of data has different modes 
    rb,wb,ab,rb+,wb+,ab+


    If we just write R,W,A, in that case it is considered as text mode.
     T is not mandatory in text mode, 
     but in byte mode B is mandatory, like RB, WB, AB, etc. 
```

```python
#  Using the `with` statement, if you open any connection in Python, 
#  it will automatically be closed when the use is done. 
# Here we don't assign it to the variable. 
# We use the alias for the open statement as f, and with reference to f we perform operations like close or truncate. 
# For instance, we create an alias f for it and f.close, f.truncate etc. 
# The connection stays open until the indentation is the same.
# s As soon as the indentation changes, the connection gets closed. 


with open("/Users/admin/Movies/test/abc.txt","r") as f:
    print(f.closed)
print(f.closed)
```
