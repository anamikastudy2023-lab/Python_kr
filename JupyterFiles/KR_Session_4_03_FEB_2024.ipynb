```python
day=input("Enter day :")
amount=int(input("Enter amount :"))

if day.lower()=="sunday" and amount>=5000:
    disocunt=(amount*10)/100
    amount=amount-disocunt
    print("after 10% discount final price will be :",amount)
else:
    print("without discount final price will be :",amount)
```

```python
day=input("Enter day :")
amount=int(input("Enter amount :"))

if day.lower()=="sunday" and amount>=5000:
    disocunt=(amount*10)/100
    amount=amount-disocunt
    print("after 10% discount final price will be :",amount)
elif day.lower()=="sunday" and amount>=4000:
    disocunt=(amount*5)/100
    amount=amount-disocunt
    print("after 5% discount final price will be :",amount)
else:
    print("without discount final price will be :",amount)
```
