```python
# parameterised query - way 1

import pymysql


try:
    con=pymysql.connect(host="localhost",user="root",password="",database="userdb133")
    cur=con.cursor()
    data=[(102,"Deepak"),(103,"Sam"),(104,"Rohan")]     #. - this data can come from anywhere at run time - user input /or file 
    sql="insert into user values (%s,%s)"        # when multiple values - parameterised value 
    cur.executemany(sql,data)
    con.commit()
except Exception as ex:
    print(ex)
finally:
    con.close()
```

```python
# parameterised query - way 2

# via user input


import pymysql


try:
    con=pymysql.connect(host="localhost",user="root",password="",database="userdb133")
    cur=con.cursor()
    id=int(input("Enter id :"))
    name=input("Enter name :")

    sql="insert into user values ('%d','%s')"%(id,name) # when multiple values - parameterised value
    cur.execute(sql)
    con.commit()
except Exception as ex:
    print(ex)
finally:
    con.close()
```
