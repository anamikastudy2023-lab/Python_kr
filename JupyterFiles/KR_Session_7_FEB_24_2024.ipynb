```python
local : inside function
global : outside function
```

```python
p=100 #global

def m1():
    q=200 #local
    print(p,q)

def m2():
    print(p)
    print(q)
    
m1()
m2()
```

```python
p=100 #global

def m1():
    global q 
    q=200
    print(p,q)

def m2():
    print(p)
    print(q)
    
m1()
m2()
```
