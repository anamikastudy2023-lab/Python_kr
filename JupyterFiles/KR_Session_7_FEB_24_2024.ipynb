```python
p=100 #global

def m1(): 
    # global q   # local variable can also be made global in this way, but rarely used, use it mostly wherever necessary. 
    q=200 #local
    print(p,q)

def m2():
    print(p)
    print(q) # ---- not a local variable so will give error, if global Q is defined in M1, then the below error will not occur. 
    
m1()
m2()
```

```python
p=100 #global

def m1():
    global q 
    q=200
    print(p,q)

def m2():
    print(p)
    print(q)
    
m1()
m2()
```
