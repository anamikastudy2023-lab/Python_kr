```python
class Person:
    def dec(self,id,name):
        self.id=id
        self.name=name
        
    def display(self):
        print(self.id)
        print(self.name)
```

```python
class Student(Person):
    
    def dec_s(self,id,name,c_name,c_fee):
        self.dec(id,name)
        self.c_name=c_name
        self.c_fee=c_fee
        
    def display_s(self):
        self.display()
        print(self.c_name)
        print(self.c_fee)
```

```python
s=Student()
```

```python
s.dec_s(101,"Amit","python",5000)
```

```python
s.display_s()
```
