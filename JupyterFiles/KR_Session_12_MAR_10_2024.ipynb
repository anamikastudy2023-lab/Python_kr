```python
# re.compile(pattern) - Compiles a regex pattern into a Pattern object
# p.finditer(string) - Returns an ITERATOR of all Match objects found
# Finds ALL occurrences of "python" in the file content
# Match object methods:
#   m.start() - Returns the START index of the match
#   m.end()   - Returns the END index of the match
#   m.group() - Returns the actual MATCHED string
#   m.span()  - Returns (start, end) as a tuple



data=open("/Users/admin/Documents/error.txt").read()

p=re.compile("python") #Pattern

# type(p). - re.pattern - it returns the type of pattern.
# Use any class; we use the object of that class. 

Compile function compiles and returns us the object of pattern class. - Used for searching and matching applications 
A search and a match are found, then compile function returns the object of the match class. 




matcher=p.finditer(data)      #returns list here   ## p.finditer(string) - Returns an ITERATOR of all Match objects found - all occurances 

for m  in matcher:
    print(m.start(),'...',m.end(),'....',m.group())
    #print(m.span())

    
```

```python
# QUANTIFIER {n} - Matches EXACTLY n occurrences of the pattern
# [ab]{2} means: match exactly 2 characters that are either "a" or "b"
# Matches: "ab", "ba", "aa", "bb", "ab", "ba" (non-overlapping)
# finditer scans left-to-right without overlapping
# Position 0-1: "ab", 2-3: "ba", 4-5: "aa", 6-7: "bb", etc.
data="abbaaabbabbaa"

p=re.compile("[ab]{2}") #Pattern



matcher=p.finditer(data)

for m  in matcher:
    print(m.start(),'...',m.end(),'....',m.group())
    #print(m.span())
```

```python
# p.fullmatch(string) - Checks if the ENTIRE string matches the pattern
# Unlike match() or search(), fullmatch() requires the WHOLE string to match
# [abc]{3,5} means: 3 to 5 characters, each being "a", "b", or "c"
# "cbb" has 3 characters (all a/b/c) -> matches! returns "found"
# "cbbx" would NOT match (contains "x" which is not in [abc])
# "cb" would NOT match (only 2 chars, minimum is 3)
data="cbb"

p=re.compile("[abc]{3,5}") #Pattern



m=p.fullmatch(data)

if m:
    print("found")
else:
    print("not found")
```

```python
# ============================================================
# REGEX QUANTIFIERS (How many times to match)
# ============================================================
# [abc]    => Matches exactly ONE character (a, b, or c)
# [abc]+   => Matches ONE or MORE characters (1 to unlimited) - greedy. atleast one of the chars a,b, or c ,  max kitna bhi ho 
# [abc]*   => Matches ZERO or MORE characters (0 to unlimited) - greedy. ek b nai h to b chaleyga , hazar b hai to b chaleyga 
# [abc]?   => Matches ZERO or ONE character (optional). atmost   - ek b nahi ya sirf inme se ek 
# [abc]{n} => Matches EXACTLY n characters
# [abc]{n,}  => Matches n or MORE characters (minimum n)
# [abc]{n,m} => Matches BETWEEN n and m characters (inclusive)
[abc] => a,b,c
[abc]+ =>
[abc]* =>
[abc]? =>
[abc]{n} =>     
[abc]{n,}    #kam se kamm 3 , jyada kitna bhi like password - atleast 8 .. uske upar kitna bhi 
[abc]{n,m}
```

```python
# Indian Phone Number Validation Rules:
# Rule 1: First digit must be 6, 7, 8, or 9 => [6-9]
# Rule 2: Remaining digits must be 0-9, total 10 digits => \d{9}
# Combined pattern: [6-9]\d{9}
1 : 6,7,8,9
2 : digit : 10
```

```python
# Indian Phone Number Validation - PROGRESSIVE pattern building:
# Step 1: [6-9][0-9]{9} - First digit 6-9, then 9 more digits
# Step 2: [6-9]\d{9} - Same but using \d shorthand for [0-9]
# Step 3: (0|\+91)[6-9]\d{9} - Must start with 0 or +91 (country code)
# Step 4: (0|\+91)?[6-9]\d{9} - 0 or +91 is OPTIONAL (? = 0 or 1)
# Step 5: With dashes format: 7234-567-899
# | means OR, () is a GROUP, ? makes group optional
# p.fullmatch() ensures the ENTIRE string must match the pattern
#data="7234567899"  ----1
data="7234-567-899"
#p=re.compile("[6-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9]") #Pattern -----1
#p=re.compile("[6-9][0-9]{9}") #Pattern is alternate of previous   -----1
#p=re.compile("[6-9]\d{9}")     #Pattern is alternate of previous   -----1
#p=re.compile("(0|\+91)[6-9]\d{9}").     | is an or symbol 
#p=re.compile("(0|\+91)?[6-9]\d{9}")
p=re.compile("(0|\+91)?[6-9]\d{3}-\d{3}-\d{3}")
m=p.fullmatch(data)

if m:
    print("found")
else:
    print("not found")
```

```python
# p.search(string) - Searches for the pattern ANYWHERE in the string but only the first occurrence anywhere in the string
# Returns the FIRST match found (not all matches)
# Unlike match(), search() doesn't require pattern at the beginning
# "is python easy to learn python" -> finds "python" at position 3-9
# Returns a Match object with span=(3, 9) showing position
# Only returns the FIRST occurrence, use finditer() for ALL
data="is python easy to learn python"

p=re.compile("python") #Pattern



m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")

```

```python
# p.sub(replacement, string) - SUBSTITUTES/REPLACES all matches
# Replaces every occurrence of the pattern with the replacement string
# re.compile("java") creates pattern to find "java"
# p.sub("python", st) replaces ALL "java" with "python"
# "java is easy to learn java" -> "python is easy to learn python"
# Similar to str.replace() but with regex pattern support
# For replacing only first N: p.sub("python", st, count=1)

# Q. String and replace it with another string. Also, we can specify which occurrence or how many occurrences we want to replace. 

st="java is easy to learn java"

p=re.compile("java") #// instead of java ve can give a pattern instead,

resp=p.sub("python",st). #- replaces all occurances 
resp=p.sub("python",st, 1) #- replaces one occurrence since we have mentioned one 


resp
```

```python
# Practical Example: CAPITALIZE first letter of lowercase lines in a file
# open("file", "r+") - Opens file for both READING and WRITING
# p.match(line) checks if line starts with lowercase letter [a-z]
# p.sub(replacement, string, count=1) - replaces only FIRST match
# line[0].upper() converts first character to uppercase
# f.truncate(0) - Clears the file content (resets file size to 0)
# f.write(st) - Writes the modified content back to the file
# f.close() - Closes the file to save changes

Q.  In the file, check wherever the first letter of the line is in lower case, converted into uppercase. 
f=open("/Users/admin/Documents/error.txt","r+")

p=re.compile("[a-z]")

st=""

for line in f:
    if p.match(line):
        line=p.sub(line[0].upper(),line,1) 
        # From the data coming in to p substitute at line 0, which is in the zero index, to uppercase, in a line and only one character. 
        print(line)
        st=st+line
    else:
        st=st+line
f.truncate(0)    #clear old data in the file 
f.write(st)     #write the updated data 
f.close()          # close it 
```

```python
# p.split(string) - SPLITS the string by the pattern matches
# Similar to str.split() but uses regex patterns as delimiters
# [|$] matches EITHER "|" or "$" as a delimiter
# "hello|this|is$kr$network" is split at each | and $ character
# Result: ['hello', 'this', 'is', 'kr', 'network']
# Useful when a string has MULTIPLE different delimiters

# for pattern based splitting of string 
st="hello|this|is$kr$network"

p=re.compile("[|$]")

resp=p.split(st)

resp
```

```python
# re.IGNORECASE (or re.I) - Flag for CASE-INSENSITIVE matching
# re.compile("python", re.IGNORECASE) matches: python, Python, PYTHON, etc.
# Without this flag, "python" would NOT match "Python"
# The flag is passed as the second argument to re.compile()
# "is Python easy to learn" -> finds "Python" (case ignored)
data="is Python easy to learn"

p=re.compile("python",re.IGNORECASE) #Pattern  ignores case sensitivity 

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```

```python
# ^ (caret) ANCHOR - Matches pattern only at the START of the string
# "^python" means: "python" must appear at the BEGINNING
# "python is easy to learn" -> starts with "python" -> found!
# "is python easy" -> does NOT start with "python" -> not found
# ^ inside [] like [^abc] means NEGATION (different meaning!)
# ^ outside [] means START OF STRING
data="python is easy to learn"

p=re.compile("^python") #Pattern

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```

```python
# $ (dollar) ANCHOR - Matches pattern only at the END of the string
# "python$" means: "python" must appear at the END
# "python is easy to learn python" -> ends with "python" -> found!
# Match is at position 24-30 (the LAST "python")
# $ is the opposite of ^ (start anchor)
data="python is easy to learn python"

p=re.compile("python$") #Pattern

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```

```python
# ============================================================
# ANCHOR vs NEGATION - Don't confuse ^ in different positions!
# ============================================================
# [^abc]  => NEGATION: matches any character EXCEPT a, b, c
# ^[abc]  => ANCHOR: string must START with a, b, or c
# [abc]$  => ANCHOR: string must END with a, b, or c
# Key: ^ INSIDE [] = negation, ^ OUTSIDE [] = start anchor
[^abc] => a,b,c
^[abc] => starts with a,b,c, searches at the beginning of the targetted string ... the first word itself
[abc]$ => endswith a,b,c
```
