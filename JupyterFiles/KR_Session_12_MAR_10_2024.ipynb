```python
f=open("/Users/admin/Documents/error.txt","r+")

p=re.compile("[a-z]")

st=""

for line in f:
    if p.match(line):
        line=p.sub(line[0].upper(),line,1)
        st=st+line
    else:
        st=st+line
f.truncate(0)
f.write(st)
f.close()
```

```python
data="is Python easy to learn"

p=re.compile("python",re.IGNORECASE) #Pattern

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```

```python
data="python is easy to learn"

p=re.compile("^python") #Pattern

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```

```python
data="python is easy to learn python"

p=re.compile("python$") #Pattern

m=p.search(data)

if m:
    print("found",m)
else:
    print("not found")
```
