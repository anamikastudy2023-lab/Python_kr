```python
class Employee:
    def __init__(self,empid,empname,salary):
        self.empid=empid
        self.empname=empname
        self.salary=salary
        
    def __mul__(self,other):
        if self.empid==other.empid:
            return self.salary*other.days
```

```python
class WorkingDays:
    def __init__(self,empid,days):
        self.empid=empid
        self.days=days
```

```python
e1=Employee(101,"Amit",567)
w1=WorkingDays(101,28)

```

```python
from abc import ABC,abstractclassmethod


class CG(ABC):
    @abstractclassmethod
    def tax(self):
        pass
```

```python
class Up(CG):
    def tax(self):
        print("16% tax")
```

```python
u=Up()
u.tax()
```

```python
import socket


s1=socket.socket()

host='127.0.0.1'
port=5000

#print(socket.gethostname())
try:
    s1.bind((host,port))
except Exception as ex:
    print(ex)

s1.listen()
print("server listening...")


while True:
       con,address=s1.accept()
       print("got connection from ,",address)
       con.send(b"hello i m TCP Server...")
       data=con.recv(1024).decode()
       print(data)
       con.close()
       break
s1.close()
```

```python
import socket


s1=socket.socket()

host='127.0.0.1'
port=5000

s1.connect((host,port))

data=s1.recv(1024).decode()
print(data)
s1.send(b"hello i m TCP client")

s1.close()
```

```python
import socket


s1=socket.socket(type=socket.SOCK_DGRAM)

host='127.0.0.1'
port=5001

#print(socket.gethostname())
try:
    s1.bind((host,port))
except Exception as ex:
    print(ex)

print("server listening...")


while True:
       data,address=s1.recvfrom(1024)
       print("got data from :",address)
       print(data.decode())
       s1.sendto(b"hello i UDP Server..",address)
       break
s1.close()
```
