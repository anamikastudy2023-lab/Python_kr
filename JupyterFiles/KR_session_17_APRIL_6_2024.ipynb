```python
import time
from threading import *


def move():
    for x in range(5):
        print("object moving..")
        time.sleep(1)

def jump():
    for x in range(5):
        print("object jumping..")
        time.sleep(1)


t1=Thread(target=move)
t2=Thread(target=jump)


t1.start()
t2.start()
```

```python
#process : 
#thread :
import time
from threading import *

l=Lock()
def msg(name):
    l.acquire()
    for x in range(5):
        print("Good Morning..",end="")
        time.sleep(1)
        print(name)
    l.release()

t1=Thread(target=msg,args=('Sachin',))
t2=Thread(target=msg,args=('Dhoni',))
t3=Thread(target=msg,args=('Virat',))


t1.start()
t2.start()
t3.start()
```

```python

import time
from threading import *
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double value :',2*n)


def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Squaers value :',n*n)

numbers=[1,2,3,4,5,6]
begintime=time.time()
t1=Thread(target=doubles,args=(numbers,))
t2=Thread(target=squares,args=(numbers,))
t1.start()
t2.start()
endtime=time.time()
print('The total time taken :',endtime-begintime)
```

```python

import time
from threading import *
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double value :',2*n)


def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Squaers value :',n*n)

numbers=[1,2,3,4,5,6]
begintime=time.time()
t1=Thread(target=doubles,args=(numbers,))
t2=Thread(target=squares,args=(numbers,))
t1.start()
t2.start()
t1.join()
t2.join()
endtime=time.time()
print('The total time taken :',endtime-begintime)
```

```python
import socket
from _thread import *

ServerSocket = socket.socket()

host = '127.0.0.1'
port = 1233
ThreadCount = 0

try:
    ServerSocket.bind((host, port))
except socket.error as e:
    print(str(e))

print('Waitiing for a Connection..')
ServerSocket.listen(5)


def threaded_client(connection):
    connection.send(str.encode('Welcome to the Server\n'))
    while True:
        data = connection.recv(2048)
        reply = 'Server Says: ' + data.decode('utf-8')
        if not data:
            break
        connection.send(str.encode(reply))
    connection.close()

while True:
    Client, address = ServerSocket.accept()
    print('Connected to: ' + address[0] + ':' + str(address[1]))
    start_new_thread(threaded_client, (Client, ))
    #t1=Thread(target=threaded_client, args=(Client, ))
    #t1.start()
    ThreadCount += 1
    print('Thread Number: ' + str(ThreadCount))
ServerSocket.close()


#sendall=>this method continues to send data from string until either all data has been sent or an error occurs



```

```python
import socket

ClientSocket = socket.socket()
host = '127.0.0.1'
port = 1233


try:
    ClientSocket.connect((host, port))
except socket.error as e:
    print(str(e))

Response = ClientSocket.recv(1024)
print(Response)
while True:
    Input = input('Say Something: ')
    ClientSocket.send(str.encode(Input))
    Response = ClientSocket.recv(1024)
    print(Response.decode('utf-8'))

ClientSocket.close()
```
