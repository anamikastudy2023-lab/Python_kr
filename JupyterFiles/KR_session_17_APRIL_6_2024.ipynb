```python
Trying move and jump without  multithreading 

import time
from threading import *


def move():
    for x in range(5):
        print("object moving..")
        time.sleep(1)

def jump():
    for x in range(5):
        print("object jumping..")
        time.sleep(1)


move()
jump()
```

```python
# Trying move and jump   with multithreading 

import time
from threading import *


def move():
    for x in range(5):
        print("object moving..")
        time.sleep(1)

def jump():
    for x in range(5):
        print("object jumping..")
        time.sleep(1)


t1=Thread(target=move)
t2=Thread(target=jump)


t1.start() #Method starts with thread, which in turn calls the target method, move. 
t2.start() #Method starts with thread, which in turn calls the target method, jump.

# python Thread is the main thread which runs the other two threads, T1 and T2. 

# now move and jump execute parallely
```

```python
# example with lock and lock acquire and release 


#process : 
#thread :
import time
from threading import *

l=Lock()    - lock is a predefined class
def msg(name):
    l.acquire()     # ---- is block me ek bar me ek hi thread ghusta hai jab tak ye block complete nahi ho jata 
    for x in range(5):
        print("Good Morning..",end="")
        time.sleep(1)
        print(name)
    l.release()

t1=Thread(target=msg,args=('Sachin',))      -converted args to tuple 
t2=Thread(target=msg,args=('Dhoni',))
t3=Thread(target=msg,args=('Virat',))


t1.start()
t2.start()
t3.start()
```

```python
# Trying to calculate time needed for - running the program  using multi-threading 


import time
from threading import *
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double value :',2*n)


def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Squaers value :',n*n)

numbers=[1,2,3,4,5,6]
begintime=time.time()
t1=Thread(target=doubles,args=(numbers,))
t2=Thread(target=squares,args=(numbers,))
t1.start()
t2.start()
endtime=time.time()
print('The total time taken :',endtime-begintime)

Here we have a problem, as there are three threads: main thread, d1 thread, d2 thread, when we are using multi-threading. 
Now, what can happen is even before T1 and T2 complete the task, the main thread might complete its task, 
which is calculation of time and printed before everything gets completed,
 which should happen at the end of T1 and T2 processing. 
 As you can see, the total time taken is calculated before t1 and t2 process or complete their calculation, which is wrong. 

 To avoid such a scenario, use join in thread. 
```

```python

 

import time
from threading import *
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double value :',2*n)


def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Squaers value :',n*n)

numbers=[1,2,3,4,5,6]
begintime=time.time()
t1=Thread(target=doubles,args=(numbers,))
t2=Thread(target=squares,args=(numbers,))
t1.start()
t2.start()
t1.join()
t2.join()
endtime=time.time()
print('The total time taken :',endtime-begintime)

T1.join and t2.join means I am joining the main thread after t1 and t2.
 The main thread will be executing only after t1 and t2 finish their calculation.
 So the correct time will be printed. 

 Read the concept of join in multi-threading in Python again. 
 Here, the main thread is depending on the processing of t1 and t2. And one thread depends on the processing of another; we use join. 
```

```python
import socket
from _thread import *

ServerSocket = socket.socket()

host = '127.0.0.1'
port = 1233
ThreadCount = 0

try:
    ServerSocket.bind((host, port))
except socket.error as e:
    print(str(e))

print('Waitiing for a Connection..')
ServerSocket.listen(5)


def threaded_client(connection):
    connection.send(str.encode('Welcome to the Server\n'))
    while True:
        data = connection.recv(2048)
        reply = 'Server Says: ' + data.decode('utf-8')
        if not data:
            break
        connection.send(str.encode(reply))
    connection.close()

while True:
    Client, address = ServerSocket.accept()
    print('Connected to: ' + address[0] + ':' + str(address[1]))
    start_new_thread(threaded_client, (Client, ))
    #t1=Thread(target=threaded_client, args=(Client, ))
    #t1.start()
    ThreadCount += 1
    print('Thread Number: ' + str(ThreadCount))
ServerSocket.close()


#sendall=>this method continues to send data from string until either all data has been sent or an error occurs



```
