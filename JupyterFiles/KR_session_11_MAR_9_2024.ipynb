```python
def checkAge(age):
    if type(age)!=int:
        raise TypeError(f"type of age must be int not {type(age)}")
    if age>18:
        print("valid")
    else:
        print("not valid")
```

```python
try:
    checkAge('23')
except TypeError as ex:
    print(ex)
```
