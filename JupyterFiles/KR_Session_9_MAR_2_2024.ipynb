```python
# os.stat(path) - Returns detailed info of file/directory status information
# Provides metadata like size, permissions, timestamps, etc.
x=os.stat("abc.txt")
# It returns a named tuple. In a named tuple, value is stored on the basis of its variable. 
# It is mostly system-generated, or by this technology we are not allowed to create an EMT tuple. It is given by the system. 

# In an unnamed tuple, the value is stored on the basis of the index. 
```

```python
# Displaying the stat_result object with all file metadata:
# st_mode = file permissions, st_size = file size in bytes,
# st_atime = last access time, st_mtime = last modified time,
# st_ctime = creation time (all in epoch/timestamp format)
x
```

```python
# Importing the stat module - provides constants for interpreting os.stat() results
# Contains permission flags like S_IRWXU (read/write/execute for user), etc.
import stat
```

```python
# os.chmod(path, mode) - Changes file permissions using symbolic constants
# S_IRWXU = Read+Write+Execute for User/Owner
# S_IRGRP = Read for Group, S_IWGRP = Write for Group
# S_IROTH = Read for Others
# The | (pipe) operator combines multiple permission flags
os.chmod("abc.txt",stat.S_IRWXU |stat.S_IRGRP |stat.S_IWGRP |stat.S_IROTH)
```

```python
# os.chmod(path, octal_mode) - Changes file permissions using octal notation
# 0o644 = Owner: read+write, Group: read, Others: read
# 0o prefix denotes an octal number in Python
os.chmod("abc.txt",0o644)
```

```python
# Importing the shutil module - provides high-level file operations
# Used for copying, moving, removing directory trees, and archiving files
import shutil
```

```python
# shutil.copy(src, dst) - Copies a file from source to destination
# Copies file content and permissions, but NOT metadata (timestamps)
# If dst is a directory, the file is copied into it with the same name
shutil.copy("newname/data.csv","dir1/")
```

```python
# shutil.copyfile(src, dst) - Copies ONLY file content (no permissions/metadata)
# Both src and dst must be complete file paths (dst cannot be a directory)
# Faster than copy() as it skips permission copying
shutil.copyfile("newname/data.csv","dir1/data.csv")
```

```python
# shutil.copy2(src, dst) - Copies file content, permissions, AND metadata
# Preserves timestamps (creation time, modification time) unlike copy()
# Most complete file copy method in shutil
shutil.copy2("newname/data.csv","dir1/")
```

```python
# shutil.copytree(src, dst) - Recursively copies an ENTIRE directory tree
# Copies all files, subdirectories, and their contents to the destination
# The destination directory must NOT already exist
shutil.copytree(".","/Users/admin/Desktop/abc")
```

```python
# shutil.rmtree(path) - Recursively DELETES an entire directory tree
# Removes the directory and ALL its contents (files + subdirectories)
# Unlike os.rmdir(), this works on non-empty directories
shutil.rmtree("newname")ftAGZA22
```

```python
# shutil.move(src, dst) - Moves a file or directory to a new location
# Works like cut+paste; can also be used to rename files/directories
shutil.move("dir1/dev.json","dir2")
```

```python
# shutil.get_archive_formats() - Returns a list of supported archive formats
# Shows available formats like zip, tar, gztar, bztar, xztar
# Mostly used for compressing files.  
shutil.get_archive_formats()
```

```python
# shutil.make_archive(base_name, format, root_dir) - Creates an archive file
# base_name = output archive path (without extension)
# format = archive type ("zip", "tar", "gztar", etc.)
# root_dir = directory to archive
shutil.make_archive("/Users/admin/Desktop/logs","zip",".")
shutil.make_archive("/Users/admin/Desktop/logs{}","zip",".")         #custom namingof files passing variable
```

```python
# shutil.unpack_archive(filename, extract_dir) - Extracts/Unzips an archive
# Extracts the archive contents into the specified directory
shutil.unpack_archive('/Users/admin/Desktop/logs.zip',"/Users/admin/Desktop/logs")
```

```python
Method	Description
shutil.copy(src, dst)	Copies file content + permissions
shutil.copyfile(src, dst)	Copies only file content
shutil.copy2(src, dst)	Copies content + permissions + metadata
shutil.copytree(src, dst)	Recursively copies entire directory tree
shutil.rmtree(path)	Recursively deletes entire directory tree
shutil.move(src, dst)	Moves file/directory (cut + paste)
shutil.get_archive_formats()	Lists supported archive formats
shutil.make_archive(...)	Creates an archive file
shutil.unpack_archive(...)	Extracts an archive
Other methods:
Method	Description
time.ctime(timestamp)	Converts Unix timestamp to readable date
oct(x.st_mode)	Converts file permissions to octal
```

```python
Method	Description
os.getcwd()	Returns the Current Working Directory as a string
os.chdir(path)	Changes the Current Working Directory
os.mkdir(name)	Creates a single new directory
os.makedirs(path)	Creates directories recursively (nested)
os.rmdir(name)	Removes a single empty directory
os.removedirs(path)	Removes nested empty directories recursively
os.remove(filepath)	Deletes a file (not a directory)
os.rename(old, new)	Renames a file or directory
os.listdir(path)	Lists all files and directories in a path
os.walk(path)	Traverses directory tree recursively
os.stat(path)	Returns detailed file metadata
os.chmod(path, mode)	Changes file permissions
os.path module methods:
Method	Description
os.path.basename(p)	Returns the filename from a full path
os.path.dirname(p)	Returns the directory portion of a path
os.path.getsize(p)	Returns file size in bytes
os.path.getctime(p)	Returns file creation time
os.path.getmtime(p)	Returns last modification time
os.path.getatime(p)	Returns last access time
os.path.join(dir, file)	Joins directory and filename into a path
os.path.exists/isfile/isdir/ismount/islink	Path validation checks
shutil module methods:
Method	Description
shutil.copy(src, dst)	Copies file content + permissions
shutil.copyfile(src, dst)	Copies only file content
shutil.copy2(src, dst)	Copies content + permissions + metadata
shutil.copytree(src, dst)	Recursively copies entire directory tree
shutil.rmtree(path)	Recursively deletes entire directory tree
shutil.move(src, dst)	Moves file/directory (cut + paste)
shutil.get_archive_formats()	Lists supported archive formats
shutil.make_archive(...)	Creates an archive file
shutil.unpack_archive(...)	Extracts an archive
Other methods:
Method	Description
time.ctime(timestamp)	Converts Unix timestamp to readable date
oct(x.st_mode)	Converts file permissions to octal
```
